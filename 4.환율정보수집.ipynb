# Iframe

- 화면상에 정보는 보이지만 실제 데이터는 다른 공간에 존재할 때 사용


```python
# iframe은 다른 공간에 존재하는 정보를 화면에 표출만 해준다
# 해결법 = 실제 데이터가 존재하는 url로 요청을 보내야한다
# iframe태그의 src안에 있는 경로를 사용하면 된다
# 대부분의 경로는 상대경로로 작성이 되어있다
# 상대경로를 절대경로와 합쳐서 url 제작
```

```python
# iframe 경로로 찾아서가야함!
res = req.get("https://finance.naver.com/marketindex/exchangeList.naver")
res.text

soup= bs(res.text,"lxml")
```

```python
title = soup.select("body > div > table > tbody > tr > td.tit > a")
money = soup.select("body > div > table > tbody > tr > td.sale")
```

```python
# 개행문자 혹은 공백 등을 지울때는 strip()을 활용

title_list = []
money_list = []
list1 = []
for i in title:
    title_list.append(i.text.strip())
    
for i in money:
    money_list.append(i.text.strip())
    
for i in range(1,59):
    list1.append(i)

title_list
money_list
```

```python
import pandas as pd
dic = {"통화명": title_list,"매매기준율":money_list, "순위":list1}
fiance = pd.DataFrame(dic)

fiance.set_index("순위",inplace=True)
fiance
```
