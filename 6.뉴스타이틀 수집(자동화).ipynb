```python
from selenium import webdriver as wb
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
```

```python
#1. 크롬 부라우저를 실행

# 1-1 네이버 메인으로 이동
driver = wb.Chrome()
driver.get("http://www.naver.com")

```

```python
# 검색창에 검색어를 입력
search = driver.find_element(By.CSS_SELECTOR,"#query")
search.send_keys("류현진")
```

```python
# 검색창에 엔터티를 입력
search_bnt = driver.find_element(By.CSS_SELECTOR,"button.btn_search")
search_bnt.click()
```

```python
# 뉴스페이지로 넘어가기
tab = driver.find_element(By.CSS_SELECTOR,"#lnb > div.lnb_group > div > ul > li:nth-child(2) > a")
tab.click()
```

```python
# 뉴스 타이틀을 수집
news = driver.find_elements(By.CSS_SELECTOR,"li> div.news_wrap.api_ani_send > div > a")


#sp_nws6 > div.news_wrap.api_ani_send > div > a
```

```python
# btn 클릭
btn = driver.find_element(By.CSS_SELECTOR,"a.btn_next")
btn.click()
```

```python
# 반복문 작성
new_list = []

for i in range(10):
    news = driver.find_elements(By.CSS_SELECTOR,"li> div.news_wrap.api_ani_send > div > a")
    for j in news:
        new_list.append(j.text)
    btn = driver.find_element(By.CSS_SELECTOR,"a.btn_next")
    btn.click()
    # time.sleep("초") = 몇초동안 코드를 멈춰라!
    # 클라이언트가 서버에게 요청을 보냈을때 사용
    # 응답받는 과정에 시간이 필요하기 때문에
    # 이미 넘겨받은 데이터(html)에서 텍스트, 내용을 추출할 때는
    # 필요하지 않다!
    time.sleep(1)
```
