```python
# 크롬드라이버 설치

# https://chromedriver.chromium.org/downloads

# 주의점 : 반드시 내가 사용중인 크롬과 같은 버전을 설치
# 현재 작업중인 폴더에 파일을 넣어준다!!
```

```python
# 라이브러리 호출
# webdriver = 크롬창 역할을 해준다!
from selenium import webdriver as wb

# 컴퓨터용 키보드 라이브러리
from selenium.webdriver.common.keys import Keys

# 크롬드라이버 최신버전 설치하는 방법
# !pip install webdriver_manager
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

# 선택자를 구분지어주는 라이브러리
from selenium.webdriver.common.by import By
```

```python
# 1. 클롬을 실행
#  wb.Chrome()을 기본으로 실행하는 명령
#  매개변수 -> 크롬드라이버의 위치(경로)를 작성
#  작성하지 않아도 되는 이유는 현재 작업중인 파일과 드라이버가 같은경로면 작성하지 않아도 된다!
driver =wb.Chrome(service=Service(ChromeDriverManager().install()))
```

```python
# 페이지를 이동
driver.get("http://www.naver.com")
```

```python
# 검색창을 찾아오기
# find_element()  --> 필요한 요소를 찾는 명령
# find_elements() --> 복수의 요소를 찾는 명령
search = driver.find_element(By.ID,"query")
```

```python
# send_key() -> 원하는 값을 전달할 때 사용
search.send_keys("오타니")
```

```python
# 엔터치기
search.send_keys(Keys.ENTER)
```

```python
# 화면 스크롤 하는 방법
body = driver.find_element(By.TAG_NAME,"body")
body.send_keys(Keys.END)
```

```python
# 페이지 뒤로 가기
# 크롬이 가지고 있는 기능이기 때문에, 태그가 아니다!
driver.back()
```

```python
# 다시한번 검색창에 검색하기
# 화면이 까빡거린다 --> 화면이 전환이 일어난다
# 컴퓨터는 모든 기억을 잃어 버린다.
# 반드시 요소를 재검색해야 한다!!!
search = driver.find_element(By.ID,"query")
search.send_keys("손흥민")
```

```python
# 검색버튼을 클릭하기
search_btn = driver.find_element(By.CLASS_NAME,"btn_search")
search_btn.click()
```

```python
# 창을 종료
driver.quit()
```

```python
import time
```

```python
title_list =[]

for i in range(0,10):
    title = driver.find_elements(By.CSS_SELECTOR,"a.news_tit")
    for j in title:
        title_list.append(j.text)
    btn = driver.find_element(By.CSS_SELECTOR,"a.btn_next")
    btn.click()
    time.sleep(2) # 2초동안 
```
